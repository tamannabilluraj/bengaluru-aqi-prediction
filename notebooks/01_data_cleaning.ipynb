```python
import pandas as pd
import numpy as np
df = pd.read_csv('../data/raw/city_day.csv')
print(df.shape)
df.head()
```

```python
print(df['City'].unique())
```

```python
df_bengaluru = df[df['City'] == 'Bengaluru'].copy()
print(df_bengaluru.shape)
```

```python
print(df_bengaluru['Date'].min())
print(df_bengaluru['Date'].max())
```

```python
print(df_bengaluru.isnull().sum())
```

```python
df_bengaluru['Date'] = pd.to_datetime(df_bengaluru['Date'])
df_bengaluru = df_bengaluru.sort_values('Date').reset_index(drop=True)
```

```python
df_bengaluru = df_bengaluru.drop(columns=['Xylene', 'Benzene', 'Toluene', 'NH3', 'PM10'])
```

```python
df_bengaluru = df_bengaluru.dropna(subset=['AQI'])
print(f"Rows remaining: {len(df_bengaluru)}")
```

```python
cols_to_fill = ['PM2.5', 'NO', 'NO2', 'NOx', 'CO', 'SO2', 'O3']
for col in cols_to_fill:
    df_bengaluru[col] = df_bengaluru[col].fillna(df_bengaluru[col].median())

print(df_bengaluru.isnull().sum())
```

```python
df_bengaluru.to_csv('../data/processed/bengaluru_clean.csv', index=False)
print("Saved!")
```
